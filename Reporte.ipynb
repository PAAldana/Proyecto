## __Exploración__:

Un precio de referencia es el valor estimado de un bien en el mercado actual, para poder estimar su valor se deben tomar en cuenta diversos cálculos,variables y fuentes de información ya que estas provocaran una fluctuación en el valor del precio. Las fuentes de datos para generar un precio de referencia para el presente proyecto son las siguientes:

* Guatecompras (Pagina de internet)
* Precios internacionales (fuentes de información de otros países en internet)
* Contratos Abiertos (Página de Guatecompras)
* Precios Antiguos (Precios almacenados en una base de datos de excel)
* Cotizaciones (Investigacion del mercado actual, almacenado en excel)

Para una correcta gestión de la información, todos los datos proveniente de las fuentes anteriormente mencionadas, se almacenaran en hojas de cálculo de excel tipo csv.
